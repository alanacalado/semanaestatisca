# Criando o gráfico de linhas

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.grid()
#plt.scatter()
plt.show()
```

# Transformando um Gráfico em Figura

- aumenta o potêncial dos comandos

```python

# 8 - polegadas
# 4 - polegadas
# # 1 polegada = 2.54 cm
# largura 8 = 20.32 cm
# altura  4 = 10.16 cm
# DPI (dots per inch, ou pontos por polegada).
# Padrão no matplotib - DPI = 100
# Largura : 8 polegadas x 100 DPI = 800 pixels.
# Altura  : 4 polegadas x 100 DPI = 400 pixels.
plt.figure(figsize=(8,4))
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.title("Imigração do Brasil para o Canadá")
plt.xlabel('Ano',fontsize=9)
plt.ylabel('Número de imigrantes',fontsize=9)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.show()
```

# Criando Subplots e adicionando rótulos

```python
plt.figure(figsize=(8,4))
# 1 polegada = 2.54
# largura 8 = 20.32 cm
# altura 4 = 10.16 cm
# DPI(dots per inch,ou pontos por polegada)
# Padrão no matplotlib - DPI = 100
# Largura : 8 polegadas x 100 DPI = 800 pixels
# Altura : 4 polegadas x 100 DPI = 400 pixels
plt.plot(df_brasil['ano'],df_brasil['imigrantes'], color='purple',marker='X')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.fill_between(df_brasil['ano'], df_brasil['imigrantes'], color='purple', alpha=0.2)
plt.yticks([500,1000,1500,2500,3000])
plt.grid(True)
plt.title("Imigração do Brasil para o Canadá", color = 'purple')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.show()
```

# Gráfico de linha e boxplot em subplots

```python
fig,axs = plt.subplots(1,2,figsize=(15,5))

# Subplot 1 : Gráfico de Linha
axs[0].plot(df_brasil['ano'],df_brasil['imigrantes'],color = 'violet',)
axs[0].set_title("Imigração do Brasíl para o Canadá  \n de 1980 a 2013", color = 'orange')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].grid(True)
axs[0].fill_between(df_brasil['ano'], df_brasil['imigrantes'], color='purple', alpha=0.2)

# Subplot 2: Gráfico Boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title("Boxplot da Imigração do Brasíl para o Canadá  \n de 1980 a 2013", color = 'orange')
axs[1].set_xlabel('Brasíl',color = 'blue')
axs[1].set_ylabel('Número de imigrantes',color = 'green')
axs[1].grid(True)


plt.show()
```

# Comparativo dos maiores paises da America Latina (imigração Canadá)

```python
# criar uma figura
fig, axs = plt.subplots(2,2,figsize=(10,6))
# wspace - Espaço Horizontal 
# hspace - Espaço Verticaç
fig.subplots_adjust(hspace=0.9,wspace=0.5)
# subplots
# Brasil
axs[0,0].plot(df.loc['Brasil',anos] ,color = 'green')
axs[0,0].set_title('Brasíl')

# Colômbia
axs[0,1].plot(df.loc['Colômbia',anos],color = 'yellow')
axs[0,1].set_title('Colômbia')

# Argentina
axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

# Peru
axs[1,1].plot(df.loc['Peru',anos],color = 'red')
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid(True)

for ax in axs.flat:
  ax.set_xlabel('Ano',fontsize=10)
  ax.set_ylabel('Números de Imigrantes',fontsize=10)

# Usando ravel() - achatar a matriz evitando o uso complexo do for
# Usando set_ylim () - definir o limite os valores.
ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)
# suptitle
fig.suptitle("Imigração na América Latina",fontsize=16)
# Customizando o Matplolib (cores, fontes, bordas)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 7))

cor_brasil = '#FFA500'  
cor_padrao = '#D3D3D3'  
cores = [cor_brasil if pais == 'Brasil' else cor_padrao for pais in america_sul_sorted.index]

bars = ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores, height=0.6)

for bar in bars:
    width = bar.get_width()
    ax.text(width + 0.02 * width, bar.get_y() + bar.get_height() / 2,
            f'{width:.0f}', va='center', ha='left', fontsize=10, color='black')

ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', fontsize=18, fontweight='bold')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

ax.xaxis.grid(False)

plt.tight_layout()

plt.show()



```
